```python
from MyFuncCollection import *
import numpy as np
from scipy import stats
```

```python
parent_dist = stats.norm(0, 1.0)
gt_dist = gentruncated(parent=parent_dist, lower = -2.0, upper = 3.0, warning = True)
tn_dist = stats.truncnorm(loc = 0, scale = 1.0, a = -2.0, b = 3.0)
```

```python
x = np.linspace(-3.0, 5.1, 1000)
plt.plot(x, gt_dist.pdf(x))
plt.plot(x, tn_dist.pdf(x))
plt.plot(x, parent_dist.pdf(x))

plt.figure()
plt.plot(x, gt_dist.cdf(x))
plt.plot(x, tn_dist.cdf(x))
plt.plot(x, parent_dist.cdf(x))

plt.figure()
plt.plot(x, gt_dist.logpdf(x))
plt.plot(x, tn_dist.logpdf(x))
plt.plot(x, parent_dist.logpdf(x))

plt.figure()
plt.plot(x, gt_dist.logcdf(x))
plt.plot(x, tn_dist.logcdf(x))
plt.plot(x, parent_dist.logcdf(x))
```

```python
plt.plot(x, gt_dist.pdf(x) - tn_dist.pdf(x))

plt.figure()
plt.plot(x, gt_dist.cdf(x) - tn_dist.cdf(x))

plt.figure()
plt.plot(x, gt_dist.logpdf(x) - tn_dist.logpdf(x))

plt.figure()
plt.plot(x, gt_dist.logcdf(x) - tn_dist.logcdf(x))

u = np.linspace(0., 1., 1000)
plt.figure()
plt.plot(x, gt_dist.ppf(u) - tn_dist.ppf(u))
```

```python
print(gt_dist.mean() - tn_dist.mean())
print(gt_dist.var() - tn_dist.var())
print(gt_dist.stats('s') - tn_dist.stats('s'))
print(gt_dist.stats('k') - tn_dist.stats('k'))
```
